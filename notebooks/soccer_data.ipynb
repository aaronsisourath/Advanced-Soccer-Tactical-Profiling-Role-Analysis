```python
import soccerdata as sd
import pandas as pd

# 1. Initialize the scraper
fbref = sd.FBref(leagues=['ENG-Premier League'], seasons='2526')

# 2. Use ONLY the valid stat types recognized by the library
stat_types = ['standard', 'shooting', 'misc']

for stat in stat_types:
    print(f"Downloading {stat} data...")
    
    # Fetch data
    df = fbref.read_player_season_stats(stat_type=stat)
    
    # 3. Clean the MultiIndex
    df.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in df.columns]
    
    # 4. Export
    filename = f"pl_2526_{stat}_stats.csv"
    df.to_csv(filename, index=True)
    print(f"Successfully saved to {filename}")

print("\nAll exports complete.")
```
