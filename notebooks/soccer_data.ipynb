```python
import soccerdata as sd
import pandas as pd

# 1. Initialize the FBref scraper
# You can add more leagues (e.g., ['ENG-Premier League', 'ESP-La Liga', 'ITA-Serie A'])
fbref = sd.FBref(leagues=['ENG-Premier League'], seasons='2526')

# 2. Define the sub-tables you need for your tactical profiling
stat_types = ['shooting', 'passing', 'possession']

for stat in stat_types:
    print(f"Downloading {stat} data...")
    
    # Fetch player-level season stats
    df = fbref.read_player_season_stats(stat_type=stat)
    
    # 3. Clean the MultiIndex (FBref data often has nested headers)
    # This flattens headers like ('Standard', 'Gls') into 'Standard_Gls'
    df.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in df.columns]
    
    # 4. Export to CSV
    filename = f"pl_2526_{stat}_stats.csv"
    df.to_csv(filename, index=True)
    print(f"Successfully saved to {filename}")

print("\nAll exports complete. You can now import these into your analysis tool.")
```
